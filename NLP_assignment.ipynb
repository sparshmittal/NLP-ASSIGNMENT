```python
#LOADING THE DATA AND CREATING SPACY OBJECT
import spacy
import textacy
nlp= spacy.load('en_core_web_lg')
file = open('AlphaGo.txt','rt')
data = file.read()
doc = nlp(data)
len(doc)
```

```python
#STEP -1 ---> SENTENCE SEGMENTATION
# READING INDIVIDUAL SENTENCES IN OUR DATA
sentences = list(doc.sents)
for sentence in sentences:
  print(sentence)
```

```python
len(sentences)
```

```python
#STEP -2  STRING TOCKENIZATION
#SPLIT OUR DATA INTO INDIVIDUAL WORDS.

for token in doc:
  print (token, token.idx)
```

```python
#PRINTING THE TOKEN AND WHETHER IT IS A STOP TOKEN OR NOT. WE WILL FILTER THE STOP WORDS IF ANY.

for token in doc:
  print (token, token.is_stop)
```

```python
#COUNTING THE MOST OCCURING WORDS IN OUR DATA.
from collections import Counter
words = [token.text for token in doc]
word_freq = Counter(words)
print(word_freq.most_common(5))
```

```python
#STEP-3 ---> PARTS OF SPEECH TAGGING(POS)
#Part of speech tagging is the process of assigning a POS tag to each token depending on its usage in the sentence. 
#POS tags are useful for assigning a syntactic category like noun or verb to each word.
for token in doc:
  print (token, token.tag_, token.pos_, spacy.explain(token.tag_))
```

```python
#Rule-Based Matching Using spaCy. # HERE SEARCHING A SINGLE WORD NOUN AND THE RESULT SHOWS ALPHAGO WHICH IS A NOUN
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
def extract_full_name(nlp_doc):
    pattern = [{'POS': 'NOUN'}]
    matcher.add('FULL_NAME', [pattern])
    matches = matcher(nlp_doc)
    for match_id, start, end in matches:
        span = nlp_doc[start:end]
        return span.text

extract_full_name(doc)
```

```python
#STEP-6 --->  NAMED ENTITY RECOGNITION
#Named Entity Recognition (NER) is the process of locating named entities in unstructured text 
#and then classifying them into pre-defined categories, such as person names, organizations, locations, monetary values, percentages, time expressions, and so on.
for entity in doc.ents:
    print(f"{entity.text} {spacy.explain(token.pos_)} ({entity.label_}) ")
```

```python
#STEP-7 --->DEPENDENCY PARSING
#Dependency parsing is the process of generating the relationship among different words of a sentence
#tocken.dep_  returns the dependency tag of a word and head.text returns the respective headword
print(f"{'Token':{8}} {'dependence':{6}} {'head text':{9}}  {'Dependency explained'} ")
for token in nlp(doc):
     print(f"{token.text:{8}} {token.dep_+' =>':{10}}   {token.head.text:{9}}  {spacy.explain(token.dep_)} ")
```

```python
#STEP -8 ---> NOUN PHRASE EXTRACTION.
#extracting noun chunks and their respective sentence from our text.
for noun_chunk in doc.noun_chunks:
    print(noun_chunk.text, "\t", noun_chunk.sent)
```

```python
#EXTRACTING SEMISTRUCTURED STATEMENTS OF OUR TOPIC ALPHAGO.
import textacy
sentences = textacy.extract.semistructured_statements(doc,entity ='AlphaGo',cue= 'be' )

for sentence in sentences:
    subject, verb, fact = sentence
    print(f" - {fact}")
```

```python
#STEP-9 ---> COREFERENCE RESOLUTION. THIS MATCHES OUR NOUN WITH THE PRONOUNS.
#AS WE CAN SEE IN THE OUTPUT ---- AlphaGo: [AlphaGo, It, It, its, AlphaGo, it, AlphaGo, it, it, it, AlphaGo, AlphaGo, AlphaGo, AlphaGo]
#THIS SHOWS OUR ORIGINAL NOUN IS ALPHAGO AND WHEN WE SEE FIRST FEW LINES OF OUR DATA, WE FIND THAT "IT" HAS BEEN USED AS A REFERENCE OF "ALPHAGO" IN SOME SENTENCES.
#EXAMPLE --- "AlphaGo" is a computer program that plays the board game Go. "It" was developed by DeepMind Technologies a subsidiary of Google.

import neuralcoref
import spacy

nlp= spacy.load('en')
neuralcoref.add_to_pipe(nlp)

file = open('AlphaGo.txt','rt')
data = file.read()
doc = nlp(data)

print(doc._.coref_clusters)

```

```python

def printPronounReferences(doc):
  print("\nPronuoun and their references")
  for token in doc:
    if token.pos_ == 'PRON' and token._.in_coref:
      for cluster in token._.coref_clusters:
        print(token.text," -> ",cluster.main.text) 

printPronounReferences(doc)
```
